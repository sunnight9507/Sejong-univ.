#### 시각화의 가치
#### 시각화에 대한 어려운 문제

- 하루에 생성되는 데이터 양
    - "하루에 25조 5천억 바이트 이상의 데이터가 생성되고, 거기서부터만 증가할 겁니다. 2020년까지 1.7명으로 추정됨MB의 데이터는 지구상의 모든 사람을 위해 매초마다 생성될 것이다."
    
- 시각화란?
    - 상징을 기하학으로 변환
    - 우리의 자연적인 인식 수단을 가장 잘 뒷받침하는 인위적인 기억 찾기
    - 인식을 증폭시키기 위해 컴퓨터에서 생성되고 상호 작용하며 시각적으로 데이터를 표현하는 방법
    
- 왜 시각화를 만드는지?
    - 질문에 대한 답변(또는 답변)
    - 의사 결정
    - 컨텍스트에서 데이터 보기
    - 메모리 확장
    - 그래픽 계산 지원
    - 패턴 찾기
    - 주장 제시 또는 스토리 전개
    - 기록 정보 : 사진, 지진데이터 등
    - 추론을 지원하기 위한 데이터 분석
        - 가설 개발 및 평가, 데이터 오류 검색, 메모리 확장, 패턴 찾기
    - 다른 사람에게 정보 전달
        - 공유 및 설득, 협업 및 수정
        
- 시각화의 가치

    - 추론을 지원하기 위한 데이터 분석

- 데이터
    - 데이터 모델 vs 개념 모델
        - 데이터 모델은 데이터에 대한 낮은 수준의 설명
            - 산술: 연산이 있는 집합
            - 예: + 및 × 연산자가 있는 정수
        - 개념 모델은 정신적 구성
            - 의미론 포함 및 지원 추론
        - 예제(Data vs Conceptual)
            - (1D floats) vs 온도 
            - (3D vector of floats) vs 공간
    - 변수 유형
        - 물리적 유형
            - 스토리지 형식별 특성
            - 기계 작동별 특성
            - 예: bool, short, int32, float, double, string...
        - 추상적 유형
            - 데이터에 대한 설명 제공
            - 방법/요소로 특징지어질 수 있음
            - 계층 구조로 구성될 수 있음
            - 예: 식물, 동물, 메타조안
    - 순서, 수량
        - 기호 : 14p
        - Nominal(라벨)
            - 과일: 사과, 오렌지...
        - Ordered(순위)
            - 육류 품질: A급, AA, AAA
        - Q-Interval : 간격 (0의 임의 위치)
            - 날짜: 2006년 1월 19일
            - 위치: (LAT 33.98, LONG -118.45)
            - 기하학적 점 직접 비교할 수 없음
            - 차이(예: 간격)만 비교할 수 있음
        - Q-Ratio : 비율(고정 0)
            - 물리적 측정: 길이, 질량, 온도... 
            - 수 및 양 
            - 기하 벡터와 같이 원점이 의미 있음
    - 상대적 데이터 모델
        - 데이터를 표로 표현(관계)
        - 각 행(투플)은 단일 레코드를 나타낸다. 
            - 각 레코드는 고정된 길이의 튜플
        - 각 열(특성)은 단일 변수를 나타낸다.
            - 각 속성은 이름과 데이터 유형 집합이다. 
        - 테이블의 스키마는 테이블(관계)의 집합이다.
        - 테이블 스키마는 이름 및 데이터 유형 집합
    - 관계형 대수
        - Data transformations (sql)
        - Projection (select) 
        - Selection (where) 
        - Sorting (order by) 
        - Aggregation (group by, sum, min, …)

            - 분석 대상 수(의존 변수)

## 5. 탐색적 데이터 분석

#### 데이터 품질, 허들
#### 통계를 사용한 시각화
#### 확인 분석

- 탐색적 데이터 분석
    - 데이터 진단
    - 그래픽 방법
    - 데이터 변환
- 통계적 모델 통합
    - 통계적 가설 검정
- 탠덤에서 그래픽 및 모델 사용

- 데이터 품질 및 사용 편의성 장애
    - 데이터 누락 : 측정값 없음, 수정됨, ...?
    - 잘못된 값: 잘못된 철자법, 특이점, ...?
    - 유형 변환 : 예: 우편 번호와 격자 번호
    - 기업 결의 : 동일한 것에 대한 다른 가치?
    - 데이터 통합 : 데이터 결합 시 작업/오류
    
- 반복 탐색
    - 탐색 프로세스
        1. 질문 해결을 위한 그래픽 구성
        2. "답변" 점검 및 새로운 질문 평가
        3. 반복!
    - 적절한 데이터 변환(예: 반전, 로그)
    - "설계변동이 아닌 데이터변동 표시"
    
- 탐색분석
    - "단순한" 데이터의 경우에도 다양한 그래픽이 통찰력을 제공할 수 있다. 다시 한 번, 그래픽의 선택을 질문하는 질문에 맞춰 조정하되, 놀라움에 대해 열린 자세를 취하십시오.
    - 그래픽을 사용하여 통계 모델의 품질을 이해하고 평가할 수 있다.
    - 모델에 대한 빠른 헌신과 검증 부족은 분석을 잘못 이끌 수 있다.
    
- 모델 통합
    - 가설 검정: 그 패턴이 우연히 발생했을 확률은 얼마인가
    - 예측: 한 개 이상의 데이터 변수가 다른 변수를 얼마나 잘 예측하는가?
    - 추상적 설명: 주어진 기능에 가장 적합한 파라미터는 무엇인가? 적합성의 장점은 무엇인가?
    - 과학적 이론: 어떤 모델이 현실을 설명할까?
    
- 통계적 의미
    - 귀무 가설을 기각하는 것이 안전하다고 간주하는 임계값(또는 타당성?)
    - p < 0.05인 경우, 일반적으로 관측된 효과나 차이가 통계적으로 유의하다고 한다.
    - 이는 관측된 데이터가 우연으로 인해 발생할 확률이 5% 미만임을 의미한다.
    - 0.05의 선택은 다소 임의적인 임계값임을 유의하십시오.
        - (R. A에 의해 선택됨. 피셔)

## 14. Networks

- 그래프 및 트리 레이아웃
    - 그래프 - 데이터 간 모델 관계 - 노드와 에지
    - 나무 - 계층 구조를 가진 그래프 - N-1 가장자리가 연결된 그래프 - 부모 및 자식 노드를 사용
- 강제 방향 배치
- 모서리 번들링
- 매트릭스 다이어그램

- 트리 시각화
    - 들여쓰기
        - 선형 목록, 삽입 인코딩 깊이
            - 모든 항목을 수직 간격의 행을 따라 배치
            - 부모/자녀 관계를 나타내는 데 사용되는 들여쓰기
            - 인터페이스에서 구성 요소로 일반적으로 사용됨
            - 공간을 위한 폭과 깊이의 경합
            - 종종 많은 스크롤이 필요함
    - 노드-링크 다이어그램
        - Nodes lines/curves으로 연결되어
            - 노드는 직선 또는 곡선 선으로 연결된 공간에 분포한다.
            - 일반적인 접근 방식은 2D 공간을 사용하여 폭과 깊이를 분리하는 것이다.
            - 계층적 지향성(예: 권한 또는 일반성)을 전달하기 위해 종종 공간이 사용됨
    - 인클로저 다이어그램
        - 엔클로저별 계층 표현
    - 계층화
        - 상대적 위치 및 정렬
        - 일반적으로 빠른 속도: O(n) 또는 O(n log n), 대화형 레이아웃

- Naive 재귀적 배치
    - 잎 개수로 하위 트리 공간을 반복적으로 분할
    - 한 차원을 따라 나무의 나침반
    - 다른 차원을 따라 깊이

- 링골드 & 틸포드의 "타이디" 배치
    - 목표: 공간을 보다 현명하게 활용하고, 밀도와 대칭을 극대화
    - 원래 2진수, Walker에 의해 일반 케이스를 커버하기 위해 확장됨
    - Buchheim 등에 의해 선형 시간 알고리즘을 달성하도록 수정

- 클러스터 덴드로그램
    - 계층적 클러스터링 알고리즘에 의해 생성된 클러스터 트리 감소
    - 라인으로 배열된 리프 노드, 내부 노드 깊이는 클러스터가 병합되는 순서/값을 나타냄
    - 직교 2-세그먼트 가장자리가 있는 순한 재귀 배치
    
- 반지름 트리 레이아웃
    - 극좌표의 노드 링크 다이어그램
    - 중심부에 루트가 있는 반지름 인코딩 깊이
    - 하위 트리에 할당된 각도 섹터(대개 순진한 재귀 레이아웃)
    - 여기에서 Reingold-Tilford 방법을 적용할 수도 있음

- 인클로저 다이어그램
    - 공간 인클로저를 이용한 암호 구조
    - 일반적으로 트라이맵으로 알려져 있음
    - 이점 - 전체 트리의 단일 뷰 제공 - 대형/소형 노드를 쉽게 찾을 수 있음
    - 문제 - 구조/깊이를 정확하게 읽기 어려움
    
- 서클 패킹 레이아웃
    - 노드는 크기 원으로 표시됨
    - 네스팅은 부모와 자식 관계를 보여준다.
    - 문제? - 비효율적인 공간 사용 - 부모 사이즈의 오해의 소지가 있는가?

- 트레맵스
    - 영역 인코딩을 통해 노드 값을 강조하는 계층 시각화
    - 리프 노드의 크기가 데이터 값에 비례하도록 파티션 2D 공간
    - 1990년 Schneiderman 등이 제안한 최초의 레이아웃 알고리즘으로, 하드 드라이브의 파일 크기 표시에 중점을 둠

- 스패닝 트리 레이아웃
    - 많은 그래프가 유용한 스패닝 트리 - 웹사이트, 소셜 네트워크
    - 그래프의 스패닝 트리에 트리 레이아웃 사용 - BFS/DFS에서 만든 트리 - 최소/최대 스패닝 트리
    - 빠른 트리 레이아웃을 통해 그래프 레이아웃을 대화형 속도로 재계산할 수 있음
    - 휴리스틱스는 레이아웃을 더욱 개선할 수 있음


- 스기야마스타일 배치
    - 역 에지를 사용하여 사이클 제거
    - 계층 계층 계층에 노드 할당
    - 누락된 도면층을 "채우기" 위해 더미 노드 생성
    - 도면층 내에 노드 배치, 에지 교차 최소화
    - 경로 가장자리 – 필요한 경우 배치 스플라인

- 계층 레이아웃 생성
    - 계층구조를 강조하는 스기야마식 레이아웃
    - 그러나 그래프의 사이클이 잘못 유도될 수 있음
    - 가장자리가 길면 근접성에 대한 인식이 저해될 수 있음


- 트리 레이아웃 - 들여쓰기/노드 링크/인클로저/계층
- 그래프 레이아웃 - 스패닝 트리 레이아웃, "스기야마" 레이아웃 - 호 다이어그램 - 강제 방향 레이아웃, 최적화 방법 - 매트릭스 다이어그램
