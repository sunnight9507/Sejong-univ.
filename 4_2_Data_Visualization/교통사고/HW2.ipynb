```python
import pandas as pd
```

```python
data_2015 = pd.read_csv('도로교통공단_전국_사망교통사고정보(2015)_20190910.csv', encoding = 'euc-kr')

data_2016 = pd.read_csv('2016년_교통사망사고정보.csv', encoding = 'euc-kr')

data_2017 = pd.read_csv('도로교통공단_전국사망교통사고_2017.csv', encoding = 'euc-kr')

data_2018 = pd.read_csv('도로교통공단_전국_사망교통사고정보(2018)_20190910.csv', encoding = 'euc-kr')
```

```python
print(data_2015.shape)
print(data_2016.shape)
print(data_2017.shape)
print(data_2018.shape)
```

```python
data_2016.columns
```

```python
data_2016.drop(columns = '발생분', inplace = True)
data_2017.drop(columns = '발생분', inplace = True)
```

```python
print(data_2015.shape)
print(data_2016.shape)
print(data_2017.shape)
print(data_2018.shape)
```

```python
data_2016.columns = data_2015.columns
data_2017.columns = data_2015.columns
```

```python
data = pd.concat([data_2015, data_2016, data_2017, data_2018])
```

```python
data.columns
```

```python
columns = ['발생년월일시분', '요일', '사망자수', '사상자수', '중상자수', '경상자수', '부상신고자수',
       '발생지시도', '발생지시군구', '사고유형_대분류', '사고유형_중분류', '사고유형','법규위반']
```

```python
data = data[columns].reset_index(drop=True)
```

```python
data['사고유형_대분류'].value_counts()
```

```python
data['사고유형_중분류'].value_counts()
```

```python
data[data['사상자수'] > 4]
```

```python
data['사고날짜'] = data['발생년월일시분'].apply(lambda x: x[:10])

data['사고난월'] = data['사고날짜'].apply(lambda x : x[2:7])

result1 = data.pivot_table(index = '사고난월', values = '사상자수', aggfunc = 'count').rename(columns = {'사상자수' : '사고횟수'})
x = data.pivot_table(index = '사고난월', values = '사상자수', aggfunc = 'sum')

result1 = pd.concat([result1, x], axis = 1)

result1.head()
```

```python
result2 = data.pivot_table(index = '사고날짜', values = '사상자수', aggfunc = 'count').rename(columns = {'사상자수' : '사고횟수'})

result2.head()
```

```python
result3 = data.pivot_table(index = '요일', values = '사상자수', aggfunc = 'count').reindex(index = ['월','화','수','목','금','토','일'])

result3
```

```python
data['발생시간'] = data['발생년월일시분'].apply(lambda x: int(x[11:13]))

result4 = data.pivot_table(index = '발생시간', values = '사상자수', aggfunc = 'count')

result4
```

```python
result5 = data.pivot_table(index = '발생지시도', values = '사상자수', aggfunc = 'count')

result5
```
